```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
lending_path = Path("./Resources/lending_data.csv")
lending_df = pd.read_csv(lending_path)

# Review the DataFrame
lending_df
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = lending_df["loan_status"]

# Separate the X variable, the features
X = lending_df.drop(columns="loan_status")
```

```python
# Review the y variable Series
y[:5]
```

```python
# Review the X variable DataFrame
X.head()
```

```python
y.value_counts()
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
logistic_regression = LogisticRegression(solver='lbfgs', random_state=1)
logistic_regression

# Fit the model using training data
LR_model = logistic_regression.fit(X_train,y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
predictions = logistic_regression.predict(X_test)
LR_predictions = pd.DataFrame({"Predictions": predictions, "Actual":y_test}).reset_index(drop=True)
LR_predictions
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
confusion_matrix(predictions,y_test)
```

```python
# Print the classification report for the model
print(classification_report(predictions,y_test))
```

# Use original data with RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# make a copy of the data to use
lending_SVM = lending_df.copy()
lending_SVM
```

```python
y2 = lending_SVM["loan_status"].values.reshape(-1,1)
y2
```

```python
X2 = lending_df.drop(columns="loan_status")
X2
```

```python
# Splitting into Train and Test sets
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=78)
```

Use the `StandardScaler` to scale the features data, remember that only `X2_train` and `X2_testing` DataFrames should be scaled.

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X2_train)
```

```python
# Scale the training data
X_train_scaled = X_scaler.transform(X2_train)
X_test_scaled = X_scaler.transform(X2_test)
```

```python
# Create the random forest classifier instance
rf_model = RandomForestClassifier(n_estimators=500, random_state=278)
```

```python
# Fit the model and use .ravel()on the "y_train" data. 
rf_model = rf_model.fit(X_train_scaled, y2_train.ravel())
```

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y2_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y2_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y2_test, predictions))
```

# Use KNeighborsClassifier

```python
from sklearn.cluster import KMeans
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
from sklearn.neighbors import KNeighborsClassifier
```

```python
lending_knn = lending_df.copy()
lending_knn
```

```python
# Create a list to store inertia values
inertia = []

# Create a list to store the values of k
k = list(range(1, 11))
```

```python
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(lending_knn)
    inertia.append(k_model.inertia_)
```

```python
# Create a Dictionary that holds the list values for k and inertia
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame using the elbow_data Dictionary
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot the DataFrame
import hvplot.pandas

df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k
)
```
